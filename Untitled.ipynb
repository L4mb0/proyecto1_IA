Tambien podemos aplicar esta misma regresion en todas las regiones utilizando un for loop. En nuestro caso, utilizaremos Ridge ya que nos da un mejor resultado.

```python
regiones = pd.DataFrame(df.groupby('region'))[0] 
```

```python
plt.title("Prediccion de Casos segun region") 
plt.xlabel("Fecha") 
plt.ylabel("# de casos")

for region in regiones:
    thisdf = df[df['region']==region].sort_values(by='date')
    thisdf = thisdf.assign (date = thisdf.date.apply (to_seconds))
    
    x_train=thisdf[['lat', 'lon', 'ubigeo', 'date']][:300]
    y_train=thisdf[['confirmed']][:300]
    
    x_test = amazonas_df[['lat', 'lon', 'ubigeo', 'date']][300:]
    y_test = amazonas_df[['confirmed']][300:].to_numpy().reshape(-1,1)
    
    reg = make_pipeline(StandardScaler(),
                 Ridge(max_iter=10000, tol=1e-3, alpha=70.0))
    reg.fit(x_train, y_train)
    pred=reg.predict(x_test)

    plt.plot(pred, label="Casos predecidos para %s" %region)
    plt.plot(y_test, label="Casos verdaderos %s" %region)
    plt.legend()
    plt.show()
    
    
```

Como podemos observar, hay grandes diferencias entre las predicciones y valores actuales de contagiados depende a la region, esto se debe principalmente a que cada region cuenta con un comportamiento difernte en cada una de las curvas (como se puede apreciar en los graficos iniciales), por ende, el valor alpha que vayamos a utilizar para una de las regiones (en nuestro caso, Amazonas) no necesariamente ayuda a representar apropiadamente el comportamiento de cada recta.

Tambien se observo que hay regiones con cambios bruscos en las ultimas fechas, y debido a que la data de entrenamiento no incluye este periodo de tiempo, las predicciones realizadas tienden un poco mas a ser erroneas.
