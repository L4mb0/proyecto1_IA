# LIMPIEZA DE DATOS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.columns
```

```python
df.tail()['confirmed']
```

```python
df=df[['region','lat','lon','date','confirmed','deaths']]
```

```python
df
```

```python
df=df.dropna().reset_index()
df
```

```python
del df['index']
```

```python
df
```

```python
import datetime as dt
df['date']=pd.to_datetime(df['date']).dt.strftime('%Y-%m-%dT%H:%M%:%SZ')
```

```python
df
```

# NON LINEAR REGRESSION
---

```python
groupedf = df.groupby('region')
```

```python
for title, group in groupedf:
    group.plot(kind='scatter',x='date',y='deaths',title=title)
```

```python
for title, group in groupedf:
    group.plot(kind='scatter',x='date',y='confirmed',title=title)
```
